# Lesson 27 - K-Means Clustering
K-means clustering is a method of vector quantization, originally from signal processing, that aims to partition n observations into k clusters in which each observation belongs to the cluster with the nearest mean (cluster centers or cluster centroid), serving as a prototype of the cluster. This results in a partitioning of the data space into Voronoi cells. k-means clustering minimizes within-cluster variances (squared Euclidean distances).

```python
import numpy as np

import pandas as pd
import matplotlib.pyplot as plt
```

# K-Means Clustering using SKLEARN
## Load the Data for Example 1

```python
df = pd.read_csv('data/kmeans_01.csv')
X = df.values
print(X.shape)
```

```python
plt.scatter(X[:,0], X[:,1], edgecolors='r')
plt.show()
```

## K-Means in Scikit-Learn

```python
from sklearn.cluster import KMeans
```

```python
kmeans_3 = KMeans(n_clusters=3)
kmeans_3.fit(X)

print(kmeans_3.labels_)

plt.scatter(X[:,0], X[:,1], edgecolors='k', c=kmeans_3.labels_)
plt.show()
```

```python
print(kmeans_3.cluster_centers_)
```

```python
pred = kmeans_3.predict([[40,35]])
print(pred)
```

```python
kmeans_5 = KMeans(n_clusters=5)
kmeans_5.fit(X)

print(kmeans_5.labels_)

plt.scatter(X[:,0], X[:,1],edgecolors='k', c=kmeans_5.labels_)
plt.show()
```
